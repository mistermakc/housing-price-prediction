## 5. Feature Transformation

### Checking the skewness of all selected numerical variables:

```python
# Highlighting the skewness of the data from highest to lowest
pd.concat([d.agg(['skew'])],axis=1).transpose().sort_values(by='skew', ascending=False)
```

```python
# Selecting only numerical variables
numerical = d.dtypes[d.dtypes != "object"].index
skewed_vars = d[numerical].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)

# Building a list of skewed numerical variables
skewness = skewed_vars[abs(skewed_vars) > 0.75]
skewed_lst = skewness.index

# Log transformations of numerical variables
for num in skewed_lst:
    d[num] = np.log2(d[num] + 1) 

# Highlighting transformed variables
pd.concat([d.agg(['skew'])],axis=1).transpose().sort_values(by='skew', ascending=False)
```

## 6. Data Preprocessing and Analysis

```python
# Data Preprocessing
X = d.drop('SalePrice', axis=1)
y = d['SalePrice']

# Initializing models
models = [
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor(random_state=42)),
    ('Random Forest', RandomForestRegressor(random_state=42)),
    ('Gradient Boosting', GradientBoostingRegressor(random_state=42))
]

# Defining cross-validation method to use
cv = KFold(n_splits=5, random_state=1, shuffle=True)
```

## 7. Results

```python
# Printing scores
for name, model in models:
    scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    mse_scores = -scores
    rmse_scores = np.sqrt(mse_scores)
    
    r2_scores = cross_val_score(model, X, y, scoring='r2', cv=cv, n_jobs=-1)

    print(f'{name}:')
    print(f'    Mean RMSE: {np.mean(rmse_scores):.2f}')
    print(f'    Std deviation: {np.std(rmse_scores):.2f}')
    print(f'    Mean R^2: {np.mean(r2_scores):.2f}\n')
```

## Interpretation

**1. Linear Regression:** The mean RMSE is now 0.35, and the standard deviation is 0.02. This suggests that, on average, the predictions of the Linear Regression model are about 0.35 units off from the actual values, and the variability of these errors is quite low. Despite these improvements, the model's R^2 score is 0.89, indicating that about 89% of the variance in the target variable can be predicted from the input features. Compared to the other models, this model is still performing relatively poorly.

**2. Decision Tree:** The mean RMSE is 0.46, and the standard deviation is 0.01. This suggests that, on average, the predictions of the Decision Tree model are 0.46 units off from the actual values, and the variability of these errors is very low. The model's R^2 score is 0.73, indicating that about 73% of the variance in the target variable can be predicted from the input features. Although this model is performing better than the Linear Regression model, it is outperformed by the other models.

**3. Random Forest:** The mean RMSE is now 0.37, and the standard deviation is 0.01. This suggests that, on average, the predictions of the Random Forest model are 0.37 units off from the actual values, and the variability of these errors is also quite low. The model's R^2 score is 0.87, indicating that about 87% of the variance in the target variable can be predicted from the input features. This model is performing better than both the Linear Regression and Decision Tree models, highlighting the strength of ensemble methods in reducing overfitting and improving prediction accuracy.

**4. Gradient Boosting:** The mean RMSE is now 0.36, and the standard deviation is 0.02. This suggests that, on average, the predictions of the Gradient Boosting model are 0.36 units off from the actual values, and the variability of these errors is also quite low. The model's R^2 score is 0.90, indicating that about 90% of the variance in the target variable can be predicted from the input features. This model is performing the best among all models in terms of the mean RMSE and R^2 score, suggesting that Gradient Boosting's approach of building new models to correct the errors of existing models is highly effective for this dataset.

## Final Recommendation

Based on the results of the models that were evaluated, the Gradient Boosting model demonstrates the best performance with the lowest root mean square error (RMSE) and the highest R^2 score. The RMSE of 0.36 implies that the predictions of the Gradient Boosting model are on average 0.36 units away from the actual values. The standard deviation of these errors is also quite low, further indicating the model's reliability. Moreover, the R^2 score of 0.90 suggests that the model can explain 90% of the variance in the target variable using the input features, which is the highest among all tested models.

Therefore, based on the data and the models tested, the Gradient Boosting model would be the recommended model for further use and development.
